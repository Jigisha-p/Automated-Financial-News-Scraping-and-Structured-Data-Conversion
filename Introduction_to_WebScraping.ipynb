```python
import requests
import bs4
```

```python
response = requests.get("https://www.moneycontrol.com/company-article/relianceindustries/news/RI")
```

```python
print(type(response))
```

```python
print(response.text)
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
print(type(soup))
```

```python
print(soup.title)
```

```python
para = soup.select('p')
```

```python
print(para)
```

```python
para_one = soup.select_one('p')
print(para_one)
```

```python
anchor = soup.find_all('a', attrs={'class': 'arial11_summ'})
print(anchor)
```

```python
for attr in anchor:
  print('Title: ', attr['title'])
  print('URL: ', attr['href'])
  print('----------------------')
```
